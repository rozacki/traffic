avg speed heathmap

```python
from pydruid.db import connect
import pandas as pd

conn = connect(host='druid', port=8082, path='/druid/v2/sql/', scheme='http')
curs = conn.cursor()
query = '''
select __time , "Avg mph" as "avg_speed", cast("Id" as integer) "site" 
from "400_sites_2016_2020" where  \'2019-01-01\'< __time and __time <\'2019-01-02\' 
and cast("Id" as integer) in (3,15)
'''
df = pd.read_sql_query(query,conn)
groupedby = df.groupby(['__time'])

#print(df["site"].unique())
print(df)
```

```python
import plotly.graph_objects as go
fig = go.Figure(data=go.Heatmap(
                   z=groupedby['avg_speed'].apply(pd.Series.tolist).tolist(),
                   x=df["site"].unique(),
                   y=df["__time"].unique(),
                   hoverongaps = False))
fig.show()
```
